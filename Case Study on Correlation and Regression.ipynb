## 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
y=df['Profit(in Thousands)']
x=df.drop(['Profit(in Thousands)','Date'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)

from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
prediction=model.predict(x_test)
evaluate = pd.DataFrame({'Actual': y_test, 'Predicted': prediction})
evaluate
```

From the table we find that the predicted values and the actual values are the same.

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error is:",mean_squared_error(y_test,prediction))
```
