# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
data['Buyer Gender'].value_counts()
```

```python
sns.set(font_scale=1.4)
data['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0,color=['red','green'])
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title('Bar Graph of Male Vs Female Buyers Participated in the Sales')
```

# 4. Find the top 5 cars based on their sales price.

```python
l = data.nlargest(5,["Sale Price"])
print(l[["Make","Model","Nickname","Sale Price"]])
```

# 5. Find the least 5 cars based on their Resell price.

```python
s = data.nsmallest(5,["Resell Price"])
print(s[["Make","Model","Nickname","Resell Price"]])
```
